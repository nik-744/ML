# Binary Logistic Regression


```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
ins_df = pd.read_csv('/content/insurance_data.csv')
```

```python
ins_df.head()
```

```python
ins_df.isna().sum()
```

```python
x,y=ins_df[['age']],ins_df['bought_insurance']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
ins_reg = linear_model.LogisticRegression()
ins_reg.fit(xtrain,ytrain)
```

```python
ypred=ins_reg.predict(xtest)
```

```python
ins_reg.score(xtest,ytest)
```

```python
ins_reg.coef_
```

```python
ins_reg.intercept_
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
hr_df = pd.read_csv('/content/HR_comma_sep.csv')
```

```python
hr_df.head()
```

```python
hr_df.isnull().sum()
```

```python
hr_df.info()
```

```python
hr_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in hr_df.columns:
  if hr_df[col].dtype == 'object':
    hr_df[col] = le.fit_transform(hr_df[col])
```

```python
hr_df.head()
```

```python
import seaborn as sns

sns.heatmap(data=hr_df.corr(),annot=True)
```

```python
#Plot bar charts showing impact of employee salaries on retention
sns.barplot(x='salary',y='left',data=hr_df)
```

```python
x,y = hr_df.drop(['left'],axis=1),hr_df['left']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
hr_reg=LogisticRegression()
hr_reg.fit(xtrain,ytrain)
```

```python
ypred=hr_reg.predict(xtest)
```

```python
hr_reg.score(xtest,ytest)
```

```python
from sklearn.metrics import accuracy_score
# accuracy_score(ytest,ypred)
print(accuracy_score(ytest,ypred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
